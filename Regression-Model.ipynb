```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression

# Read the dataset from Excel
data = pd.read_excel('distance_fuel_consumption.xlsx')

# Assuming the first column is the independent variable (distance) and the second column is the dependent variable (fuel consumption)
X = data.iloc[:, 0].values.reshape(-1, 1)  # Distance (input)
y = data.iloc[:, 1].values  # Fuel Consumption (output)

# Create the linear regression model
model = LinearRegression()

# Train the model
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Plot the data points
plt.scatter(X, y, color='black', label='Actual data')

# Plot the regression line
plt.plot(X, y_pred, color='red', linewidth=2, label='Regression line')

# Add labels and title
plt.xlabel('Distance (miles)')
plt.ylabel('Fuel Consumption (gallons)')
plt.title('Linear Regression: Distance vs Fuel Consumption')
plt.legend()

# Show the plot
plt.show()

# Calculate R-squared value
r_squared = model.score(X, y)
print("R-squared:", r_squared)

# Get user input for distance
user_input_distance = float(input("Enter the distance to predict the fuel consumption (in miles): "))

# Predict fuel consumption for the given distance
predicted_fuel_consumption = model.predict([[user_input_distance]])

# Print the predicted fuel consumption
print(f'Predicted fuel consumption for {user_input_distance} miles is {predicted_fuel_consumption[0]} gallons')
```
